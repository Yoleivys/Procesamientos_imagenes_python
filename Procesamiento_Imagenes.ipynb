## Imagenes digitales
Podríamos definir una imagen como una función bidimensional f(x,y ) donde x e y
son las **coordenadas espaciales**, y el valor de f en cualquier par de coordenadas
(x,y) es la intensidad de la imagen en dicho punto.
Una imagen puede ser continua con respecto a x e y, y también en intensidad
(imagen analógica). Convertir una imagen a formato digital requiere que tanto las
coordenadas como la intensidad sean digitalizadas. Digitalizar las coordenadas se
llama **muestrear**, mientras que digitalizar la intensidad se denomina
**cuantización**. Entonces, cuando todas las cantidades son discretas, llamamos a la
imagen una imagen digital.


## Creando matrices y visualizandolas como imagenes 

Las imágenes se codifican como matrices. En particular, las imágenes de intensidad o escala de grises se codifican como una matriz de dos dimensiones, donde cada número representa la intensidad de un pixel.
Pero eso significa que cualquiera de estas matrices que generamos se puede visualizar como una matriz. Para visualizar imágenes, usamos el módulo pyplot de la librería matplotlib

```python
import matplotlib.pyplot as plt   # para poder visualizar las imagenes 
import numpy as np                # para poder trabajar con matrices

#plt.rcParams['image.cmap'] = 'gray' #configuracion necesaria de pyplot para ver las imagenes en escala de grises.
                                    #No siempre es necesaria 
```

La función que nos permite visualizar matrices es **imshow** del modulo pyplot, que invocaremos como **plt.imshow**, y recibe la imagen como parámetro. Debido a que pyplot intenta ajustar los colores automaticamente de una imagen, vamos a pasarle como parámetros también vmin=0,vmax=1.

### Imagen Negra

```python
#tamaño de las matrices a visualizar
size=(5,6) #  Número de pixeles de la imagenes 5x6 = 30 

# Una matriz de ceros. 
imagen_negra = np.zeros(size)

#visualizamos la matriz
#Se ve como una imagen negra, ya que todos los elementos (pixeles) tienen intensidad 0
#plt.imshow(imagen_negra,vmin=0,vmax=1)
plt.imshow(imagen_negra, cmap='gray')
# (es necesario indicar vmin y vmax para que pyplot sepa que el minimo es 0 y el maximo 1)
# (solo imagenes escala de grises)
```

### Imagen Blanca

```python

# Una matriz de unos. 
imagen_blanca = np.ones(size)

#visualizamos la matriz
#Se ve como una imagen blanca, ya que todos los elementos (pixeles) tienen intensidad 1
plt.imshow(imagen_blanca,vmin=0,vmax=1)

#creamos otra figura para mostrar la imagen (sino el proximo imshow sobreescribe al anterior)
plt.figure()
```

### Imagen Gris

```python

# Una matriz con valor 0.5 en todos sus elementos 
imagen_gris = np.ones(size)*0.5

#visualizamos la matriz
#Se ve como una imagen gris, ya que todos los elementos (pixeles) tienen intensidad 0.5
plt.imshow(imagen_gris,vmin=0,vmax=1)
```

Clasificación

Extracción/identificación de patrones

## Lectura, visualización y escritura de imágenes.

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import io
# imagen a blanco y negro 

Chris_Gris =io.imread("Chris_Gris.jpg")/255.0 # imread lee las imagenes con los pixeles codificados como enteros 
# en el rango 0-255. Por eso la convertimos a flotante y en el rango 0-1

print("- Dimensiones de la imagen:")
print(Chris_Gris.shape) # tamaño de la imagen o matriz 

plt.imshow(Chris_Gris,vmin=0,vmax=1) # imagen a blanco y negro 

# imagen a color 
plt.figure()
Chris_rgb=io.imread("Chris.jpg")/255.0 

plt.imshow(Chris_rgb)# no es necesario el vmin/vmax para imágenes a color

print("Dimensiones de la imagen:")
print(Chris_rgb.shape)
```

### Ver canales individualmente 
Podemos ver cada canal individualmente, como si el canal fuera una imagen en escala de grises, para ver la intensidad de cada canal.

```python
plt.imshow(Chris_rgb[:,:,0],vmin=0,vmax=1)
plt.title("Canal Rojo")
plt.figure()
plt.imshow(Chris_rgb[:,:,1],vmin=0,vmax=1)
plt.title("Canal Verde")
plt.figure()
plt.imshow(Chris_rgb[:,:,2],vmin=0,vmax=1)
plt.title("Canal Azul")
```

Podemos ver que el canal rojo es el más activo, ya que la imagen contiene mucha piel.

Otra manera de ver la información de cada canal es quitar la información de los otros canales. Por ejemplo, para ver el canal rojo, ponemos en 0 los canales azul y verde.

```python
Chris_red=np.copy(Chris_rgb) # creo una copia de la imagen para preservar la original
Chris_red[:,:,1]=0
Chris_red[:,:,2]=0
plt.title("Chris_ canal rojo")
plt.imshow(Chris_red)
```

```python
Chris_green=np.copy(Chris_rgb) # creo una copia de la imagen para preservar la original
Chris_green[:,:,0]=0
Chris_green[:,:,2]=0
plt.title("Chris_ sin canal azul")
plt.imshow(Chris_green)
```

```python
Chris_red_green=np.copy(Chris_rgb) # creo una copia de la imagen para preservar la original
Chris_red_green[:,:,2]=0
plt.title("Chris_ sin canal azul")
plt.imshow(Chris_red_green)
```

# Combinando colores con filas y columnas

```python
Chris_color = Chris_rgb.copy()
# Remove green (1) & blue (2) from top-left corner.
Chris_color [:50, :50, 1:] = 0
# Remove red (0) & blue (2) from bottom-right corner.
Chris_color [-50:, -50:, (0, 2)] = 0
plt.imshow(Chris_color );
```

### Guardar imágenes
Guardar imágenes es tan simple como cargarlas con la función imsave del módulo skimage.io.

```python
io.imsave("Chris_generado.png",Chris_red_green)
```
